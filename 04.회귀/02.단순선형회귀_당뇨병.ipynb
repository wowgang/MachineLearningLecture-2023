### 단순 선형회귀 - 당뇨병 사례

```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

- 당뇨병 데이터셋 탐색

```python
from sklearn.datasets import load_diabetes 
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
df.head()
```

```python
df.describe()       # 분포가 N(0, 0.0476)로  이미 정규화되어 있음
# -2.511817e-19	  => 0이나 마찬가지이므로 0
# 표준편차 보면 4.76e-02 로 거의 비슷함 정규화 되어있음을 확인할수있다.
```

```python
print(diabetes.DESCR)
```

- 10개의 feature와 target간의 상관관계도 

```python
# 한줄에 5개씪
fig, axs = plt.subplots(2, 5, figsize=(16, 8))
for index, feature in enumerate(df.columns[:-1]):
    row, col = index // 5, index % 5
    sns.regplot(x=feature, y='target', data=df, ax=axs[row][col])
```

- 10 개 feature vs. target

```python
from sklearn.linear_model import LinearRegression 
lr = LinearRegression()
```

```python
# 결정계수(R 제곱)는 회귀 모델이 데이터에 얼마나 잘 맞는지를 나타내는 통계량입니다. 
# 예측 값과 실제 값 사이의 상관 관계를 측정합니다. 
# 0에서 1 사이의 값으로, 값이 높을수록 모델이 데이터에 더 잘 맞습니다. 
# 더 큰 값일수록 모델이 종속 변수의 변동을 더 잘 설명한다는 의미
# 예를 들어, R 제곱이 0.80이면 모델이 데이터의 변동의 80%를 설명합니다.
# R 제곱은 회귀 모델의 적합도를 평가하는 데 유용한 척도
```

```python
# 각 feature와 target의 선형회귀를 했을 때의 R squared value
# 위에 그래프 맞는지 값 확인해보기
for feature in df.columns[:-1]:
    X = df[feature].values.reshape(-1,1)       # 1차원이니까 2차원로 바꿔줘야함
    lr.fit(X, diabetes.target)
    score = lr.score(X, diabetes.target)
    print(f'{feature}: {score:.4f}')
# 기울기의 값을 볼게 아니라 score 값을 보고 회귀선의 얼마나 모여있느냐(밀집)
```

#### BMI vs target

- 훈련/테스트 데이터셋 분리 (test_size=0.1)
    - 회귀에서 분리는 좀 그렇다심 굳이 느낌? 

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(
    df.bmi.values.reshape(-1,1), diabetes.target, test_size=0.1, random_state=2023
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# 가중치(계수 W), 바이어스(X 편향=절편)
lr.coef_, lr.intercept_
# 950 * bmi + 151
```

```python
# R squared 
lr.score(X_test, y_test)
```

```python
# R squared, MSE 
from sklearn.metrics import r2_score, mean_squared_error 
pred = lr.predict(X_test)
r2_score(y_test, pred), mean_squared_error(y_test, pred)
# 0.29 중의 설명력이 있구나...
# 4598은 두값이 있어야 차이를 알수 있음
# MSE: 두 값의 차이의 제곱의 평균을 계산해줌
```

- 시각화

```python
plt.figure(figsize=(6,4))
plt.scatter(X_train, y_train)
plt.grid()
plt.xlabel('BMI'), plt.ylabel('target')
plt.title('BMI vs. target', fontsize=15)
plt.show()
```

```python
# 직선그리기
xs = np.array([-0.1, 0.18])
ys = np.dot(lr.coef_[0], xs) + lr.intercept_
ys
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X_train, y_train,label='train dataset')
plt.scatter(X_test, y_test, marker='^', label='test dataset')
plt.plot(xs, ys, 'r-', lw=3) # 'r-' 선의 색상을 빨간색('r')으로, 선의 스타일을 실선('-')으로
plt.grid(), plt.legend()
plt.xlabel('BMI'), plt.ylabel('target')
plt.title('BMI vs. target', fontsize=15)
plt.show()
```

##### BP vs target

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(
    df.bp.values.reshape(-1,1), diabetes.target, test_size=0.1, random_state=2023
)
lr.fit(X_train2, y_train2)
pred2 = lr.predict(X_test2)
```

```python
r2_score(y_test2, pred2), mean_squared_error(y_test2, pred2)
```

```python
# R^2 / MSE
# bp = 0.2365 / 4949.6
# bmi = 0.2906 / 4598.9
# r2 bmi가 좋음
# MSE는 bmi 가 좋은데 숫자가 작음  / MSE는 숫가자 작은게 좋음 
# GridSearchCV로 하려면
# scoring='mean_squared_erro'로 할경우 neg를 붙여줌 음수 / 음수를 붙여주면 작은값이 큰값이니까
# scoring 값이 커지는 방향으로 학습을 하기 때문에
# 정확도 / 정밀도/ 재현도? 값이 커야 좋음
```
