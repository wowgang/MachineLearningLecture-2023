### K-Means Algorutgm - Iris

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris 
from sklearn.preprocessing import StandardScaler 
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
iris_std = StandardScaler().fit_transform(iris.data)
pca = PCA(2)    # pca = PCA(n_components=2) # 몇차원으로 축소할건지 말해줘야함
iris_pca = pca.fit_transform(iris_std)
```

##### 1. 군집화(Clustering)

```python
from sklearn.cluster import KMeans 
KMeans().get_params()
```

```python
# verbose=1 화면에 표시 해주는 param
# n_init= 값 써줘야함  초기 클러스터 중심을 설정하는 방법을 자동으로 선택
kmeans = KMeans(n_clusters=3, random_state=2023, n_init='auto')
kmeans.fit(iris_std)    # 비지도 학습으로 y값을 제공하지 않음
```

```python
# 학습 결과
kmeans.labels_
```

```python
# 핛급 결과 넘버링은 집단이 큰 순서부터 부여
# 군집화후 이렇게 나눠짐을 확인
# 가장 많이 나온순으로 0,1,2 / 답을 알려준적 없으니 기존에 하던 target값과 다름 
# 그래서 lables_의 값과  다름을 알기
np.unique(kmeans.labels_, return_counts=True)
```

##### 2. 시각화

```python
df = pd.DataFrame(iris_pca, columns=['PC1', 'PC2'])
df['target'] = iris.target
df['cluster'] = kmeans.labels_
df.head(54).tail()
```

```python
# 데이터 분포
df.groupby(['target', 'cluster'])[['PC1']].count()
```

```python
# PCA 주성분 분석
# 3개로 군집화한 결과
markers = ['^', 's', 'o']   

plt.figure(figsize=(5,3))
for i , marker in enumerate(markers):
    x_data = df[df.cluster == i]['PC1']
    y_data = df[df.cluster == i]['PC2']
    plt.scatter(x_data, y_data, marker=marker, label='cluster'+str(i))
    plt.scatter(x_data.mean(), y_data.mean(), marker='*', c='red', s=100)
plt.legend()
plt.xlabel('PC1'), plt.ylabel('PC2')
plt.title("3 Clustering result")    
plt.show()
```

- PCA Original Data와 비교

```python
plt.figure(figsize=(10,3))
for k, column in enumerate(['target','cluster']):
    plt.subplot(1, 2, k+1) #  subplot 함수는 그림의 행(row), 열(column), 현재 서브플롯의 인덱스(index)를 지정
    for i , marker in enumerate(markers):
        x_data = df[df[column] == i]['PC1']
        y_data = df[df[column] == i]['PC2']
        if k == 0:
            plt.scatter(x_data, y_data, marker=marker, label=iris.target_names[i])
        else:
            plt.scatter(x_data, y_data, marker=marker, label='cluster'+str(i))
            plt.scatter(x_data.mean(), y_data.mean(), marker='*', c='red', s=100)
            
    plt.legend()
    plt.xlabel('PC1'), plt.ylabel('PC2')
    if k == 0:
        plt.title("Original data")    
    else:
        plt.title("Clustering result")    
plt.show()
```
